```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
train=pd.read_csv('train.csv')
test=pd.read_csv('test.csv')
```

```python
submission=pd.read_csv('sample_submission.csv')
```

```python
trainy=train['label']
trainx=train.iloc[:,1:785]
trainx.head()
```

```python
train1=StandardScaler().fit_transform(trainx)
test1=StandardScaler().fit_transform(test)
```

```python
trainx1=pd.DataFrame(train1)
testx1=pd.DataFrame(test1)
```

```python
from matplotlib import pyplot as plt

def viewDigits(whichSeries, threshold=None):
    whichSeries = np.array(whichSeries, dtype='float')
    if threshold is not None:
        whichSeries = (whichSeries > threshold).astype(int)
    plt.imshow(whichSeries.reshape((28,28)), cmap='gray')

viewDigits(trainx1.iloc[3])
```

```python
from sklearn.svm import SVC
clf = SVC(kernel='linear')
clf.fit(trainx1,trainy)
y_pred = clf.predict(testx1)
```

```python
submission.head()
```

```python
submission['Label']=y_pred
```

```python
submission.head()
submission.to_csv('Submission1.csv',index=False)
```
